```python
import re
import json
from bs4 import BeautifulSoup
```

```python
# source generated with "get links" util
with open("../data/pretty.html") as f:
    source = f.read()
```

```python
soup = BeautifulSoup(source,'lxml')
```

```python
# bounds of data regions in raw source - likely to change over time - inspect source and adjust
data_bounds = re.compile(r"\"\{\\\"data\\\".*?(?=\;(\r\n|\r|\n)window)")
```

```python
data_sections = [i.group() for i in data_bounds.finditer(source)]
```

```python
# 3 data sections appear to exist. The 3rd one contains the data we're interested in.
len(data_sections)
```
