```python
test_url = "http://www.quora.com/q/euauidtynxwaufug/How-to-eliminate-shoe-odor"
```

```python
words = set(nltk.corpus.words.words())
```

```python
body = requests.get(test_url).text
```

```python
def clean_sent(body):
    return " ".join(w for w in nltk.wordpunct_tokenize(body) \
     if w.lower() in words or not w.isalpha())
```
